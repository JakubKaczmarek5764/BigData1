# 1. Liczba zachowoań dla każdego kraju

```python
import pandas as pd

with open('lmr1/output', encoding='utf-8') as f:
    lines = f.readlines()

data = []
for line in lines:
    parts = line.strip().split()
    country = ' '.join(parts[:-1])
    value = int(parts[-1])
    data.append({'country': country, 'value': value})

# Utwórz DataFrame
df = pd.DataFrame(data)
df
```

# 4. Podobieństwo kosinusowe przebiegu pandemii pomiędzy krajami

```python
with open('lmr4/part-00000', encoding='utf-8') as f:
    lines = f.readlines()

data = []
for line in lines:
    parts = line.strip().split('\t')
    countries = parts[0].split(' - ')
    value = float(parts[1])
    data.append({'country1': countries[0], 'country2': countries[1], 'similarity': value})

df = pd.DataFrame(data)
df
```
